```python
# DATA MINING CLASS
# Supervised Learning
# 1. Classification
#     a). Logistic Regression
#     b) CART(Classification and Regression Techniques) = Random Forest, Ada Boost, Decision Trees, XG Boost
#     c) SVM Classifier
# 2. Regresssion
#     a) Linear Regression
#     b) CART
#     c) SVM Rgressor
# Unsupersised Leanring
# 1. K-Means
# 2. DBSCAN
# 3. Hierarchical 

# Problem for hte day: Customer Churn with Logistic Regression
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# creating a data frame
churndata = pd.read_csv('C:/Users/Bina/Desktop/Machine-Learning/Supervised Learning/ChurnData.csv', delimiter = ',')
churndata.head()
```

```python
#for categorical data, its important to have variable catalogues
# to investigate whether the customer will churn or not.
churndata.head() #displays the first 10 rows
```

```python
# Data Pre-processing and selection step
# assingning a subset: narrowing it down: using the chosen headers
churndata = churndata[['tenure', 'age','address','income','ed','employ','equip','callcard','wireless','churn']]
# churn it the target variable and converting teh type to int
churndata['churn'] = churndata['churn'].astype('int')
churndata.head()
```

```python
churndata.shape
```

```python
# defining x and y for datset
x = np.asarray(churndata[['tenure', 'age','address','income','ed','employ','equip']])
x[0:5]
```

```python
y = np.asarray(churndata['churn'])
y[0:5]
```

```python
# normalizing the datset
# normalization is done so that
# 1. help the algorithm converge faster - helps the 
x = preprocessing.StandardScaler().fit(x).transform(x)
x[0:5]
```

```python
# splitting data into training and testing
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=50) 
print("Train set:", x_train.shape, y_train.shape) 
print("Test set:", x_test.shape,y_test.shape)
```

```python
# building our model using scikit-learn: Logistic Regression
# in scikit-learn logistic regression, it supports
# 1. regularization - used to solve overfitting problems
# 2. C parameter - indicates the inverse of regu;arization strength which must be a positive float.
# Smaller values specify a stronger regularization.
# training the model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(x_train,y_train)
# the algorihtm now learns or trains from the .fit()
LR #trained object - has knowledge based on teh dataset we have 
```

```python
# testing the model
# predict using the test set
# yhat are the predicted values form the x_test
yhat = LR.predict(x_test)
yhat

# to get teh class probabilities
yhat_prob = LR.predict_proba(x_test)
yhat_prob
```

```python
# check if the predicted churn is it the actual truth
# if the prediction is 1 and the truth is 0, then the model is wrong
# confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,yhat,labels=[1,0]))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import itertools
import matplotlib.pyplot as plt
import numpy as np

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    """
    The function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float')/ cm.sum(axis=1)[:,np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')
    print(cm)
    
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i,j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i,j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.show()

# Assuming y_test and yhat are defined
cm = confusion_matrix(y_test, yhat)
plot_confusion_matrix(cm, classes=['Churn=Yes', 'Churn=No'], normalize=False)

# Print classification report
print(classification_report(y_test, yhat, labels=[1, 0]))
```
