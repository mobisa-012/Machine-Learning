# 1 INTRODUCTION
## 1.1 Background
#### Primary education is the foundation of Kenya's education system and a main engine of national growth. While the government has made efforts towards expansion of access and quality, disparities in infrastructure, teacher distribution, and resources between regions prevail. This data set provides Kenya's primary schools' overall indicators such as enrollment levels, facility conditions, staffing ratios, and geographic reach. These factors can be utilized to study systemic problems and make policy decisions based on evidence. As education is one of the primary drivers of socioeconomic movement, these inequalities are key information in facilitating equitable learning outcomes. The statistics also give details on how sponsorship types of schools (government, religious, private) affect resource distribution.
## 1.2 Research Problem
#### Kenya's primary schools undergo deep inequities in resource distribution that impact education quality. Schools suffer from overcrowded classrooms, unsanitary facilities, and inadequate teachers, primarily within marginalized regions. Aggregate national statistics may indicate rising trends, yet localized gaps are poorly studied. Such concealed inequalities will be uncovered in this analysis through examination of: 
     1. spatial patterns of infrastructure shortages,
     2. provincial variations in pupil-teacher ratios, and 
     3. gender variation in facility allocation. 
#### Without such in-depth analysis, resource allocation can perpetuate existing imbalances, with disadvantaged groups being underserved.

#### Objectives
    1. To assess adequacy of physical facilities (classrooms, toilets) in relation to numbers on roll. 
    2. To analyze patterns of teacher distribution and areas of severe staffing shortages.
    3. To analyze gender differences in facility allocation and enrolment by school type.

#### From the above list, the sponsors data is messy categorical data and needs to be cleanded and standardized. I'll do this by mappring them.
#### By doing this:
        1. I reduce the noise from tyops and inconsitencies
        2. I avoid overcounting similar sponsors like RELIGIOUS ORGANIZATION AND RELIGIOUS ORGANISATION
        3. It will enable me better aggregation, visualization and modelling

```python
# cleanign teh blanks and the dashes,-
data['School_Sponsor'] = data['School_Sponsor'].replace(r'^\s*$', 'UNKNOWN', regex=True)
data['School_Sponsor'] = data['School_Sponsor'].replace('-', 'UNKNOWN')

# I'll now standardize using mappi
sponsor_mapping = {
    'RELIGIOUS ORGANISATION': 'RELIGIOUS ORGANIZATION',
    'PRIVATE INDIVIDUAL/ORGANISATION': 'PRIVATE INDIVIDUAL/ORGANIZATION',
    'CENTRAL GOVERNMET/DEB': 'CENTRAL GOVERNMENT/DEB',
    'CENTRAL GORVERNMENT/DEB': 'CENTRAL GOVERNMENT/DEB',
    'CENTRAL  GOVERNMENT/DEB': 'CENTRAL GOVERNMENT/DEB',
    'CENTRAL GOVERNMENT': 'CENTRAL GOVERNMENT/DEB',
    'LOCAL GORVERNMENT AUTHORITY': 'LOCAL GOVERNMENT AUTHORITY',
    'LOCAL AUTHORITY': 'LOCAL GOVERNMENT AUTHORITY',
    '': 'UNKNOWN',
    '-': 'UNKNOWN'
}

# applying the above mapping
data['School_Sponsor'] = data['School_Sponsor'].replace(sponsor_mapping)
sponsors = data['School_Sponsor'].value_counts()
print(sponsors)
```

```python
# Group by Sponsor AND Province
province_stats = data.groupby(['School_Sponsor', 'Province']).agg({
    'Pupil_Teacher_Ratio': 'median',
    'Pupil_Classroom_Ratio': 'median'
}).sort_values(['Province', 'Pupil_Teacher_Ratio'])

# Visualize top 5 provinces
plt.figure(figsize=(12, 6))
sns.barplot(data=data.nlargest(5, 'Total_Enrollment'), 
            x='Province', y='Pupil_Teacher_Ratio', 
            hue='School_Sponsor', palette='viridis')
plt.title('Pupil-Teacher Ratio by Sponsor (Top 5 Provinces)')
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy import stats

# ANOVA test for Pupil-Teacher Ratio across sponsors
f_val, p_val = stats.f_oneway(
    data[data['School_Sponsor'] == 'PRIVATE']['Pupil_Teacher_Ratio'],
    data[data['School_Sponsor'] == 'RELIGIOUS']['Pupil_Teacher_Ratio'],
    data[data['School_Sponsor'] == 'GOVERNMENT']['Pupil_Teacher_Ratio']
)
print(f"ANOVA p-value: {p_val:.4f}")  # p < 0.05 confirms significant differences
```

```python
# Staffing analysis
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Pupil-Teacher Ratio distribution
sns.histplot(data['Pupil_Teacher_Ratio'], bins=30, kde=True, 
            ax=axes[0,0], color='purple')
axes[0,0].axvline(x=data['Pupil_Teacher_Ratio'].median(), 
                 color='red', linestyle='--')
axes[0,0].set_title('Distribution of Pupil-Teacher Ratios', fontsize=14)
axes[0,0].set_xlabel('Pupils per Teacher')

# Teacher type comparison
teacher_types = data[['GOK_TSC_Teachers_Male', 'GOK_TSC_Teachers_Female',
                    'Authority_Teachers_Male', 'Authority_Teachers_Female',
                    'PTA_BOG_Teachers_Male', 'PTA_BOG_Teachers_Female']].sum()
teacher_types.plot(kind='bar', ax=axes[0,1], color=['blue', 'pink']*3)
axes[0,1].set_title('Teacher Types by Gender', fontsize=14)
axes[0,1].set_ylabel('Total Number of Teachers')

# Pupil-Teacher Ratio by Province
sns.boxplot(y='Province', x='Pupil_Teacher_Ratio', data=data,
           ax=axes[1,0], palette='viridis')
axes[1,0].set_title('Pupil-Teacher Ratio by Province', fontsize=14)
axes[1,0].set_xlabel('Pupils per Teacher')

# Teaching vs Non-Teaching Staff
staff_ratio = data[['Num_Teaching_Staff', 'Num_Non_Teaching_Staff']].mean()
staff_ratio.plot(kind='bar', ax=axes[1,1], color=['green', 'orange'])
axes[1,1].set_title('Average Teaching vs Non-Teaching Staff', fontsize=14)
axes[1,1].set_ylabel('Average Number of Staff')

plt.tight_layout()
plt.show()
```

```python
# Enrollment analysis
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Total Enrollment distribution
sns.histplot(data['Total_Enrollment'], bins=30, kde=True, 
            ax=axes[0,0], color='darkblue')
axes[0,0].set_title('Distribution of Total Enrollment', fontsize=14)
axes[0,0].set_xlabel('Number of Students')

# Gender Enrollment comparison
gender_enrollment = data[['Num_Boys', 'Num_Girls']].sum()
gender_enrollment.plot(kind='bar', ax=axes[0,1], color=['blue', 'pink'])
axes[0,1].set_title('Total Enrollment by Gender', fontsize=14)
axes[0,1].set_ylabel('Number of Students')

# Enrollment by School Type
sns.boxplot(y='Student_Gender_Type', x='Total_Enrollment', data=data,
           ax=axes[1,0], palette='pastel')
axes[1,0].set_title('Enrollment Distribution by School Gender Type', fontsize=14)
axes[1,0].set_xlabel('Number of Students')

# Enrollment vs Classroom Ratio
sns.scatterplot(x='Total_Enrollment', y='Pupil_Classroom_Ratio', 
               data=data, ax=axes[1,1], hue='School_Sponsor', 
               alpha=0.6, palette='Set2')
axes[1,1].set_title('Enrollment vs Classroom Ratio', fontsize=14)
axes[1,1].set_xlabel('Total Enrollment')
axes[1,1].set_ylabel('Pupils per Classroom')

plt.tight_layout()
plt.show()
```
